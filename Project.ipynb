```python
def find_cat_num(df):
    global cat_data
    global num_data
    cat_data=[]
    num_data=[]
    for i in list(df.columns):
        if len(df[i].unique())/len(df[i]) <0.2:
            cat_data.append(i)
        else:
            num_data.append(i)
```

```python
find_cat_num(df)
print(cat_data,num_data,sep='\n')
```
